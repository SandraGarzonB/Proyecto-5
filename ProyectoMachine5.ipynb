```python
from scipy.spatial.distance import mahalanobis
```

```python
AB_num= AB.select_dtypes(exclude='category')
```

```python
mahal_distances = []

for row in AB_num.to_numpy():
  mahal_distances.append(mahalanobis(row, AB_num.mean(), np.linalg.inv(np.cov(AB_num.values.T))))
```

```python
k = AB_num.shape[0]

AB_num['mahal_distances'] = mahal_distances
AB_num['p_value'] = 1 - chi2.cdf(AB_num['mahal_distances'], k-1)
```

```python
AB_num= AB_num[AB_num['p_value'] > 0.05]
```

```python
AB_num.drop(columns=['mahal_distances','p_value'],inplace=True)
```
